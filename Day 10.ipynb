# Dynamic Plotting

# ### Visualization effect of viscosity, flowrate and permeability on flow profile

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from ipywidgets import interactive
from IPython.display import display
```

```python
def flowprofile(k,mu,q):
    re = 3000
    rw = 0.5
    
    r = np.linspace(rw,re,500)
    pe = 4000
    
    B = 1
    h = 30
    
    P = pe - (141.2*q*mu*B*(np.log(re/r))/(k*h))
    
    y_min = P[np.where(r==rw)]
    plt.figure(figsize=(10,8))
    plt.plot(r,P,linewidth=4)
    plt.axhline(y_min,linewidth=3,color='red')
    plt.ylim(0,5000)
    plt.xlabel('r(ft)')
    plt.ylabel('P(r), Psi')
    
    plt.title('Reservoir Pressure Profile')
    
    plt.grid(True)
    
    return r,P
```

```python
interactive_flow_profile = interactive(flowprofile,k = (20,1000), mu =[5,10,15],q = (100,2000))
```

```python
display(interactive_flow_profile)
```

```python
def circle(r):
    x = np.linspace(-r,r,1000)
    y_p = (r**2 - x**2)**0.5
    y_n = -y_p
    
    plt.figure(figsize = (8,8))
    plt.plot(x,y_p,color = 'red')
    plt.plot(x,y_n,color ='red')
    plt.ylim(-12,12)
    plt.xlim(-12,12)
    plt.grid()
    plt.title(f"Circle of radius {r}")
    
    return x, y_p,y_n
```

```python
interactive_circle = interactive(circle,r = (1,10))
```
